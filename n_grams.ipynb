### Data

### Frecuency

```python
frec_tweets, frec_news, frec_blogs = (Counter(), Counter(), Counter())

for i in tweets:
    frec_tweets.update(w for w in i.split())
for i in news:
    frec_news.update(w for w in i.split())
for i in blogs:
    frec_blogs.update(w for w in i.split())
```

### N-grams

```python
def calc_2gram(data):
    
    n_gram = []
    
    for tmp_data in data:
        last = ""
        for wn in tmp_data.split():
            if last == "":
                last = wn
                continue
                
            n_gram.append((last, wn))
            last = wn

    return n_gram

def calc_3gram(data):
    
    n_gram = []
    
    for tmp_data in data:
        last = ""
        p_last = ""
        for wn in tmp_data.split():
            if p_last == "":
                p_last = wn
                continue
            elif last == "":
                last = p_last
                p_last = wn
                continue
                
            n_gram.append((last, p_last, wn))
            last = p_last
            p_last = wn

    return n_gram
```

```python
# subsection
porcentage = 0.2
length = len(multi_media)
cut = int(length*porcentage)

subsection = multi_media[:cut]
```

```python
n2_gram = calc_2gram(subsection)
n3_gram = calc_3gram(subsection)
```
